```python
df["test_name"] = df["test_name"].str.strip()

df["result_flag"] = df["result_flag"].str.strip()
```

```python
df = df.drop_duplicates()
```

```python
df["test_date"] = pd.to_datetime(
    df["test_date"],
    format="mixed",
    dayfirst=True
)
```

```python
df["result_flag"] = df["result_flag"].str.title()

df["test_name"] = df["test_name"].str.title()
```

```python
df["test_value"] = df["test_value"].fillna(
    df["test_value"].median()
)
```

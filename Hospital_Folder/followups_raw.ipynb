```python
followups = followups.drop_duplicates()
```

```python
followups['followup_due_date'] = pd.to_datetime(
    followups['followup_due_date'],
    format='mixed',
    dayfirst=True
)
```

```python
text_cols = [
    'followup_completed',
    'followup_mode',
    'followup_outcome'
]

for col in text_cols:
    followups[col] = followups[col].str.strip()
```

```python
followups['followup_completed'] = followups[
    'followup_completed'
].str.title()

followups['followup_mode'] = followups[
    'followup_mode'
].str.title()

followups['followup_outcome'] = followups[
    'followup_outcome'
].str.title()
```

```python
import pandas as pd
import numpy as np
```

```python
departments = pd.read_csv("D:\OneDrive\Documents\Downloads\Hospital_Project\departments.csv")
```

```python
departments = departments.drop_duplicates()
```

```python
text_cols = [
    'department_name',
    'care_complexity'
]

for col in text_cols:
    departments[col] = departments[col].str.strip()
```

```python
departments['department_name'] = departments[
    'department_name'
].str.title()

departments['care_complexity'] = departments[
    'care_complexity'
].str.title()
```

```python
departments['care_complexity'].unique()
```

```python
departments.info()
```

```python
departments.to_csv(
    "cleaned_departments.csv",
    index=False
)
```

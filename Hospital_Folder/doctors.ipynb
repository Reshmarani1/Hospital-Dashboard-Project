```python
doctors = doctors.drop_duplicates()
```

```python
text_cols = [
    'doctor_name',
    'specialization',
    'doctor_grade'
]

for col in text_cols:
    doctors[col] = doctors[col].str.strip()
```

```python
doctors['doctor_name'] = doctors[
    'doctor_name'
].str.title()

doctors['specialization'] = doctors[
    'specialization'
].str.title()

doctors['doctor_grade'] = doctors[
    'doctor_grade'
].str.title()
```

```python
doctors.to_csv("cleaned_doctors.csv",index=False)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:\OneDrive\Documents\Downloads\Hospital_Project\pharmacy_orders_raw.csv")
```

```python
text_cols = [
    "medicine_name",
    "pharmacy_type"
]

for col in text_cols:
    df[col] = df[col].str.strip()
```

```python
len(df)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
len(df)
```

```python
df.dtypes
```

```python
df["quantity"] = df["quantity"].fillna(
    df["quantity"].median()
)
```

```python
df["quantity"] = df["quantity"].astype(int)
```

```python
df["unit_price"] = df["unit_price"].fillna(
    df["unit_price"].median()
)
```

```python
df["line_amount"] = df["line_amount"].fillna(
    df["line_amount"].median()
)
```

```python
df.loc[
    df["quantity"] <= 0,
    "quantity"
] = np.nan
```

```python
df.loc[
    df["line_amount"] <= 0,
    "line_amount"
] = np.nan
```

```python
df.loc[
    df["unit_price"] <= 0,
    "unit_price"
] = np.nan
```

```python
df["line_amount"] = df["line_amount"].fillna(
    df["line_amount"].median()
)
```

```python
df["pharmacy_type"] = (
    df["pharmacy_type"]
    .str.strip()
    .str.title()
)
```

```python
df["pharmacy_type"].unique()
```

```python
df["pharmacy_type"] = df["pharmacy_type"].replace({
    "Ip Pharmacy": "IP Pharmacy",
    "Op Pharmacy": "OP Pharmacy"
})
```

```python
df[
    round(df["quantity"] * df["unit_price"], 2)
    != round(df["line_amount"], 2)
]
```

```python
df["line_amount"] = (
    df["quantity"] * df["unit_price"]
).round(2)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:\OneDrive\Documents\Downloads\Hospital_Project\patients_raw.csv")
```

```python
text_cols = [
    "patient_code",
    "patient_name",
    "gender",
    "city",
    "primary_chronic_condition",
    "insurance_plan",
    "income_segment"
]

for col in text_cols:
    df[col] = df[col].str.strip()
```

```python
len(df)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.dtypes
```

```python
df["registration_date"] = pd.to_datetime(
    df["registration_date"],
    format="mixed",
    dayfirst=True
)
```

```python
df["age"] = df["age"].fillna(
    df["age"].median()
)
```

```python
df["age"] = df["age"].astype(int)
```

```python
df.dtypes
```

```python
df["city"] = df["city"].fillna("Unknown")
```

```python
df["primary_chronic_condition"] = (
    df["primary_chronic_condition"]
    .fillna("Unknown")
)
```

```python
df["gender"] = df["gender"].str.title()
```

```python
df["insurance_plan"] = (
    df["insurance_plan"].str.title()
)
```

```python
df["income_segment"] = (
    df["income_segment"].str.title()
)
```

```python
df["gender"].unique()
```

```python
df["income_segment"].unique()
```

```python
df["gender"] = df["gender"].replace({
    "M": "Male",
    "F": "Female",
    "Femle": "Female"
})
```

```python
df["gender"].unique()
```

```python
df[df["age"] < 0]
```

```python
df[df["age"] > 120]
```

```python
df["age"] = pd.to_numeric(
    df["age"],
    errors="coerce"
)
```

```python
df.loc[
    (df["age"] <= 0) | (df["age"] >= 110),
    "age"
] = np.nan
```

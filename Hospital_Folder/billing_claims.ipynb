```python
billing = billing.drop_duplicates()
```

```python
billing['bill_date'] = pd.to_datetime(
    billing['bill_date'],
    format='mixed',
    dayfirst=True
)
```

```python
amount_cols = [
    'gross_amount',
    'discount_amount',
    'insurance_paid',
    'patient_paid'
]
```

```python
for col in amount_cols:
    billing[col] = billing[col].fillna(
        billing[col].median()
    )
```

```python
text_cols = [
    'payer_type',
    'claim_status'
]

for col in text_cols:
    billing[col] = billing[col].str.strip()
```
